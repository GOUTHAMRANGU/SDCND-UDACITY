# TASK : 
TO FIND THE LANE LINES ON ROAD

# PIPELINE :

1. Import image or video into the workspace.
2. Convert to grayscale.
3. Blur the image using Gaussian blur with right kernel size.
4. Mark Canny edges on the Gaussian_blurred_image.
5. Create ROI with only lanes contained and mask the frame with ROI.
6. Apply Hough transform over the ROI_CANNY_IMAGE with parameters set to required accuracy.
7. Draw line segments(edges) using co-ordinate point pairs resulted from hough transform
8. Average over the respective lines to create left and right lanes
9. Add/Combuine the line image with the original frame to check the level of accuracy of annotations.


```python
# import all the required libraries
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
import math
import os
%matplotlib inline
```

```python
def read_image(image_path):
    return mpimg.imread(image_path)
test_images = [read_image('test_images/' + i) for i in os.listdir('test_images/')]
```

```python
#define some required functions
def slope(x1,y1,x2,y2):
    m= (y2-y1)/(x2-x1)
    if math.isnan(m):
        return
    else:
        return m
#DEFINE A FUNCTION THAT ONLY KEEPS REQUIRED SLOPES  : "m_to_use" return "slope"
def mc_to_use(m):
    m_find=[]    
    if len(m)==1:
        return m[0]
    m_mean = np.mean(m)
    m_std = np.std(m)
    for slope in m:
        if slope - m_mean < 1.5*m_std:
            m_find.append(slope)
    if not m_find:
        m_find=m
    slope = np.mean(m_find)
    return slope



def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    return cv2.addWeighted(initial_img, α, img, β, λ)
    
```

```python
def process(image):
    gray = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
    kernel_size = 5
    blur_gray = cv2.GaussianBlur(gray,(kernel_size,kernel_size),0)
    low_thres = 50
    high_thres = 150
    edges = cv2.Canny(blur_gray, low_thres, high_thres)
    mask = np.zeros_like(edges)
    ignore_mask_color = 255
    imshape = image.shape
    vertices = np.array([[(20,imshape[0]),(415, 330), (520,330), (imshape[1]-20,imshape[0])]], dtype=np.int32)
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    masked_edges = cv2.bitwise_and(edges, mask)
    rho = 2 # distance resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 45     # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 40 #minimum number of pixels making up a line
    max_line_gap = 100   # maximum gap in pixels between connectable line segments
    line_image = np.copy(image)*0 # creating a blank to draw lines on
    created = np.copy(image)*0
    lines = cv2.HoughLinesP(masked_edges, rho, theta, threshold, np.array([]),
                            min_line_length, max_line_gap)
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(line_image,(x1,y1),(x2,y2),(255,0,0),2)
    #initialize lists for respective labels such as left abscissa, left ordinate, their corresponding slopes
    left_p_x= []
    left_p_y=[]
    m_left=[]
    right_p_x = []
    right_p_y = []
    m_right= []
    for line in lines:
        for x1,y1,x2,y2 in line:
            m=slope(x1,y1,x2,y2)
            #if slope>0 append the co-ordinates into left labelled lists
            if m>0:
                m_left.append(m)
                left_p_x.append(x1)
                left_p_y.append(y1)
                left_p_x.append(x2)
                left_p_y.append(y2)
            #if slope<0 append the co-ordinates into right labelled lists
            if m<0:
                m_right.append(m)
                right_p_x.append(x1)
                right_p_y.append(y1)
                right_p_x.append(x2)
                right_p_y.append(y2)
    #print(left_p_x,left_p_y)
    #print(right_p_x,right_p_y)
    
# cefine a function that averages among left lines and right lines respectively but also draws them continuously
#The idea employed here is as follows:
# The slope is averaged and the slopes that correspond to the respective direction but are aligned with error>1.5*standard deviation are ignored
# Min of the ordinate of the two lane points are extracted and using the slope and the point x-intercept is found
# cv2.line is used with arguments x-intercept and the min point.
    def create(cx,cy,dx,dy):
        k2y,l2y = 540, 540
        i= cy.index(min(cy))
        k1x,k1y = cx[i],cy[i]
        m1= mc_to_use(m_left)
        k2x= int(k1x + (k2y-k1y)/m1)
        j= dy.index(min(dy))
        l1x,l1y = dx[j],dy[j]
        m2= mc_to_use(m_right)
        l2x= int(l1x + (l2y-l1y)/m2)
        #print('right',(k1x,k1y),(k2x,k2y))
        #print('left',(l1x,l1y),(l2x,l2y))
        cv2.line(created, (k1x,k1y),(k2x,k2y),(0,255,0),10)
        cv2.line(created, (l1x,l1y),(l2x,l2y), (0,255,0),10)
        

# Create a "color" binary image to combine with line image
    wh = create(left_p_x,left_p_y,right_p_x,right_p_y)
    color_edges = np.dstack((edges, edges, edges))
    clone= np.dstack((edges,edges,edges))



# Draw the lines on the edge image
    lines_edges = cv2.addWeighted(color_edges, 0.8, line_image, 1, 0) 
    lin_edg = cv2.addWeighted(clone, 0.8, created, 1, 0)
    
    result = weighted_img(created,image)

    plt.subplot(2,2,1)
    plt.imshow(lines_edges)
    plt.subplot(2,2,2)
    plt.imshow(lin_edg)
    plt.subplot(2,2,3)
    plt.imshow(result)
    plt.subplot(2,2,4)
    plt.imshow(created)
    return result

process(test_images[0])
 
```

```python
# Import everything needed to edit/save/watch video clips
from moviepy.editor import VideoFileClip
from IPython.display import HTML
```

```python
def process_image(image):
    """Puts image through pipeline and returns 3-channel image for processing video below."""
    result = process(image)
    #print(result.shape)
    return result
```

# REFLECTION:
1. Current pipeline: In the current pipe line the slopes are averaged for the respective lane short lines and a new     line is created with the x-intercept the topmost point in the image that corresponds to the respective lanes
2. Trail methods used: Linear regression method, np.polyfit(), spliting the whole data-set into co-ordinate pairs     and joining the adjacent points.
3. Potential shortcomings: The main difference between the example media file and to the one that is processed is      that, in the example file the lanes are very stable. This method canot be used on roads where there is              illumination difference, and also curved roads. This method cannot be used in turnings as the output from the      demonstrated method is a line but not a curve.
4. Possible improvements: Using polyfit effectively over the hough points to generate curves, reduce the              disturbance in the detected lanes. Improve image resolution to detect lanes with great comfort although it comes 
   at a good computation cost.
