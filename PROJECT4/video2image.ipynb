```python
import cv2
vidcap = cv2.VideoCapture('challenge_video.mp4')
outdir = 'video2image0challenge/'
success,image = vidcap.read()
count = 0
success = True
while success:
  success,image = vidcap.read()
  print ('Read a new frame: ', success)
  cv2.imwrite(outdir+"frame%d.jpg" % count, image)     # save frame as JPEG file
  count += 1
```

```python
vidcap.release()
cv2.destroyAllWindows()
```
